```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv("cars_raw.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df[df["Price"] == "Not Priced"]
```

```python
df["Price"] = df["Price"].astype(str)
df["Price"] = df["Price"].str.replace("$","",regex=False).str.replace(",","")
df["Price"] = pd.to_numeric(df["Price"],errors="coerce")
```

```python
df["Price"].isna().sum()
```

```python
df = df.dropna(subset=["Price"])
df.reset_index(drop=True)
```

```python
df["Price"].isna().sum()
```

```python
df["Price"] = df["Price"].astype("Int64")
```

```python
df["Price"].head()
```

```python
df["Used/New"].value_counts()
```

```python
summary = []
for col in df.columns:
    col_type = df[col].dtype
    if pd.api.types.is_numeric_dtype(df[col]):
        summary.append([
            col, str(col_type), df[col].min(), df[col].max(), "Numeric"
        ])
    else:
        summary.append([
            col, str(col_type), None, None, f"Categorical ({df[col].nunique()} unique)"
        ])

summary_df = pd.DataFrame(summary, columns=["Column", "Data Type", "Min", "Max", "Category Info"])
summary_df.head(32)
```

```python
for col in df.columns:
    print(f"\nColumn: {col}")
    
    if pd.api.types.is_numeric_dtype(df[col]):
        col_data = df[col].dropna()
        print(f"  Mean: {col_data.mean():.2f}")
        print(f"  Median: {col_data.median():.2f}")
        print(f"  Variance: {col_data.var():.2f}")
        print(f"  Standard Deviation: {col_data.std():.2f}")
        
    else:
        value_counts = df[col].value_counts()
        percentages = df[col].value_counts(normalize=True) * 100
        result = pd.DataFrame({'Frequency': value_counts, 'Percentage (%)': percentages.round(2)})
        print(result.head(10))
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_cols.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            linewidths=0.5, linecolor='gray', square=True, cbar_kws={'shrink': 0.75})

plt.title('🔍 Correlation Heatmap of Numerical Features', fontsize=18, pad=20)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['DealType'].value_counts(dropna=False).plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Distribution of DealType (Including Missing)', fontsize=14)
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='SellerType', data=df, palette='viridis')
plt.title('Frequency of Seller Types', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Price', data=df, estimator='mean', ci=None, marker='o')
plt.title('Average Car Price Over the Years', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
